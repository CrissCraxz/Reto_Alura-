```python
tiendas = [tienda, tienda2, tienda3, tienda4]

for df in tiendas:

    df.rename(columns=columnas, inplace=True)

    for campo, diccionario in reemplazos.items():
        if campo in df.columns:
            df[campo] = df[campo].replace(diccionario)


    if 'Producto' in df.columns:
        df['Producto'] = (
            df['Producto']
            .apply(limpiar_texto)
            .apply(estandarizar)
            .replace(productos)
        )

```

#1. Análisis de facturación



```python

tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']


ingresos_cop = [df['Precio'].sum() for df in tiendas]

for nombre, cop in zip(nombres_tiendas, ingresos_cop):
    print(f"{nombre}: {cop:,.2f} COP")

print(tienda.columns.tolist())

```

```python
ingresos_tiendas = [tienda['Precio'].sum(), tienda2['Precio'].sum(), tienda3['Precio'].sum(), tienda4['Precio'].sum()]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(8,5))
plt.bar(nombres_tiendas, ingresos_tiendas, color='steelblue', edgecolor='black')
plt.title('Facturación Total por Tienda')
plt.ylabel('Pesos Colombianos')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

```

# 2. Ventas por categoría

```python
for i, tienda_df in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    categoria_count = tienda_df['Categoria del Producto'].value_counts()
    plt.figure(figsize=(8, 5))
    categoria_count.plot(kind='barh', color='mediumaquamarine', edgecolor='black')
    plt.title(f'Categorías más vendidas - Tienda {i}')
    plt.xlabel('Cantidad de Productos Vendidos')
    plt.tight_layout()
    plt.show()
```

# 3. Calificación promedio de la tienda


```python
calificaciones = [
    tienda['Calificacion'].mean(),
    tienda2['Calificacion'].mean(),
    tienda3['Calificacion'].mean(),
    tienda4['Calificacion'].mean()
]

plt.figure(figsize=(8,5))
plt.bar(nombres_tiendas, calificaciones, color='gold', edgecolor='black')
plt.title('Calificación Promedio por Tienda')
plt.ylabel('Promedio de Calificación')
plt.ylim(0, 5)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
for i, df in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    producto_count = df['Producto'].value_counts()
    mas_vendido = producto_count.idxmax()
    menos_vendido = producto_count.idxmin()
    print(f"Tienda {i} -> Más vendido: {mas_vendido} ({producto_count.max()} ventas), Menos vendido: {menos_vendido} ({producto_count.min()} ventas)")
```

# 5. Envío promedio por tienda

```python
envio_promedio = [
    tienda['Precio del Costo de Envio'].mean(),
    tienda2['Precio del Costo de Envio'].mean(),
    tienda3['Precio del Costo de Envio'].mean(),
    tienda4['Precio del Costo de Envio'].mean()
]

plt.figure(figsize=(8,5))
plt.bar(nombres_tiendas, envio_promedio, color='salmon', edgecolor='black')
plt.title('Costo Promedio de Envío por Tienda')
plt.ylabel('Promedio Envío (COP)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#Metodo de pago mas usado#

```python
metodo_pago_total = pd.concat([tienda, tienda2, tienda3, tienda4])
metodo_count = metodo_pago_total['Metodo de Pago'].value_counts()

plt.figure(figsize=(10,6))
metodo_count.plot(kind='bar', color='orchid', edgecolor='black')
plt.title('Frecuencia de Uso por Método de Pago')
plt.xlabel('Método de Pago')
plt.ylabel('Cantidad de Transacciones')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#Ubicacion geografica#
